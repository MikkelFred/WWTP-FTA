```r
library(igraph)
library(readxl)
library(ggplot2)
```

```r
edges_concept <- read_xlsx("Vandværk/Edges3.xlsx")
vertices_concept <- read_xlsx("Vandværk/Vertices3.xlsx")
```

```r
edges_concept.df <- as.data.frame(edges_concept)
nodes_concept.df <- as.data.frame(vertices_concept)
```

```r
g1 <- graph_from_data_frame(d = edges_concept.df, vertices = nodes_concept.df, directed = FALSE)
m1 <- layout_nicely(g1)
plot(g1, vertex.label.color = "black", layout = m1)
w1 <- E(g1)$Strength
g.b <- betweenness(g1, directed = FALSE, weights = w1, normalized = TRUE)
#hist(g.b, breaks = 100, xlab = "Vertices", col = "lightblue", border = "pink")
```

```r
plot <- plot(g1, 
     edge.color = 'black',
     vertex.size = sqrt(g.b)+10,
     edge.width = w1,
     layout = layout_nicely(g1))
```

```r
barplot(sort(g.b, decreasing=TRUE), las=2, cex.names=0.6)
```

```r
barplot(sort(degree(g1, normalized = FALSE), decreasing=TRUE), las=2, cex.names=0.6)
```
